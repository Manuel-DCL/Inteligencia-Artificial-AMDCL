```python
#Compuerta logica AND
def AND():
  a, b = Get()
  x=[a,b] #Peso
  w=[1,1] #Peso
  s=-1 #Numero fijo
  for i in range(0,2):

    print((s+x[i]*w[i]),"=",s,"+", x[i], "*", w[i]) #tester
    s=s+x[i]*w[i]
  print(s)

  y=1 if s>0 else 0
  print("!xxxxxxxxxxxxxxxxxxxxxxxxx!")
  print(x[0],'AND',x[1],'=',y)
  print("!xxxxxxxxxxxxxxxxxxxxxxxxx!\n")

#Compuerta logica OR
def OR():
  a, b = Get()
  x=[a,b] #Peso
  w=[1,1] #Peso
  s=0 #Numero fijo
  for i in range(0,2):

    print((s+x[i]*w[i]),"=",s,"+", x[i], "*", w[i]) #tester
    s=s+x[i]*w[i]
    print(s)

  y=1 if s>0 else 0
  print("!xxxxxxxxxxxxxxxxxxxxxxxxx!")
  print(x[0],'OR',x[1],'=',y)
  print("!xxxxxxxxxxxxxxxxxxxxxxxxx!\n")

#Optencion de valores para la operacion
def Get():
  print("Menciona el valor de x (0-1)")
  x = int(input())
  print("Menciona el valor de x2 (0-1)")
  y = int(input())
  return x, y

#Menu
def menu():
  while True:
    print("---------------------------------------------------------------------------")
    print("Selecciona la compuerta <1-2> selecciona cualquier otro numero para salir")
    print("1° AND")
    print("2° OR")
    print("3°.. Salir")
    print("---------------------------------------------------------------------------")
    op = int(input())
    if op == 1:
      AND()
    else:
      if op == 2:
        OR()
      else:
        break

menu()



```
